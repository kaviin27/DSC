## JSON example, with string

+ demonstrates creation of normalized dataframes (tables) from nested json string
+ source: http://pandas.pydata.org/pandas-docs/stable/io.html#normalization

```python
# define json string
data = [{'state': 'Florida', 
         'shortname': 'FL',
         'info': {'governor': 'Rick Scott'},
         'counties': [{'name': 'Dade', 'population': 12345},
                      {'name': 'Broward', 'population': 40000},
                      {'name': 'Palm Beach', 'population': 60000}]},
        {'state': 'Ohio',
         'shortname': 'OH',
         'info': {'governor': 'John Kasich'},
         'counties': [{'name': 'Summit', 'population': 1234},
                      {'name': 'Cuyahoga', 'population': 1337}]}]
```

```python
# use normalization to create tables from nested element
json_normalize(data, 'counties')
```

```python
# further populate tables created from nested element
json_normalize(data, 'counties', ['state', 'shortname', ['info', 'governor']])
```

## JSON exercise

Using data in file 'data/world_bank_projects.json' and the techniques demonstrated above,
1. Find the 10 countries with most projects
2. Find the top 10 major project themes (using column 'mjtheme_namecode')
3. In 2. above you will notice that some entries have only the code and the name is missing. Create a dataframe with the missing names filled in.

```python
#Import json data amd storing it as a dataframe
df = pd.read_json('data/world_bank_projects.json')
df.head()
```

```python
#cleaning the data. In this case specific column () for the excercise
# Work around code to get unique values and dropping value pairs with name ==''
prjt_series = df['mjtheme_namecode'].apply(pd.Series).stack().reset_index(drop=True)
mask = prjt_series.apply(lambda x: x['name']!='')
uniq_val = prjt_series[mask].value_counts()
```

```python
#Getting the values of the index to list and acquiring a dict where value of 'code'  os the key and value of 'name' is the value
uniq_val = pd.DataFrame(uniq_val)
temp_list = uniq_val.index.to_list()
val_dict = {temp_list[i]['code'] : temp_list[i]['name'] for i in range(len(temp_list))}
val_dict
```

```python
# Replacing missing 'name' values in the column using the val_dict as reference.
def replace_vals(col):
    #print(col)
    for element in col:
        for i in range(len(element)):
            if(element[i]['name']==''):
                element[i]['name'] = val_dict[element[i]['code']]
replace_vals(df.mjtheme_namecode)
```

```python
#Displaying top 10 countries with most world bank projects
df.countryname.value_counts().head(10)
```

```python
# Bar plot of the data
df.countryname.value_counts().head(10).plot(kind='bar')
plt.ylabel('# Projects')
plt.ylim(0, 24)
plt.show()
```

```python
#Displaying Major project themes
df['mjtheme_namecode'].apply(pd.Series).stack().reset_index(drop=True).value_counts().head(10)
```

```python
# Bar plot of the data
df['mjtheme_namecode'].apply(pd.Series).stack().reset_index(drop=True).value_counts().head(10).plot(kind='bar')
plt.ylabel('Count')
#plt.xticks(rotation=85)
plt.ylim(0, 260)
plt.show()
```
